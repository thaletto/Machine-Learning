**Train-Test-Spilit**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(input, target, test_size=0.20)
```

**SVM**

```python
from sklearn import svm
```

```python
svm_model=svm.SVC(kernel='linear')
```

```python
svm_model.fit(X_train,y_train)
```

**Decision Tree**

```python
from sklearn.tree import DecisionTreeClassifier as dtc
```

```python
dt_model=dtc(criterion='entropy')
```

```python
dt_model.fit(X_train,y_train)
```

```python
from sklearn import tree
```

```python
graph = tree.plot_tree(dt_model)
```

**Random Forest**

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model = RandomForestClassifier(n_estimators=10,class_weight='balanced')
```

```python
rf_model.fit(X_train,y_train)
```

**Performance Analysis**

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import plot_confusion_matrix
from sklearn.metrics import classification_report
```

```python
svm_pred=svm_model.predict(X_test)
dt_pred=dt_model.predict(X_test)
rf_pred=rf_model.predict(X_test)
```

```python
accuracy_score(y_test,svm_pred)
```

```python
accuracy_score(y_test,dt_pred)
```

```python
accuracy_score(y_test,rf_pred)
```

```python
print(classification_report(y_test,svm_pred))
```

```python
print(classification_report(y_test,dt_pred))
```

```python
print(classification_report(y_test,rf_pred))
```

```python
plot_confusion_matrix(svm_model,X_test,y_test)
```

```python
plot_confusion_matrix(dt_model,X_test,y_test)
```

```python
plot_confusion_matrix(dt_model,X_test,y_test)
```

SVM is better because of higher accuracy but Random forest can have higher accuracy if n_estimators is increased
